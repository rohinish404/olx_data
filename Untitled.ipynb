```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('results.csv')
```

```python
data
```

```python
df_new = data.copy()
df_new
```

```python
price = []
for i in range (0,len(df_new['price'])):
    str = df_new['price'][i]
    str = str.replace('₹','')
    str = str.replace(' ','')
    str = str.replace(',','')
    
    price.append(int(str))
price
```

```python
df_new['new_price'] = price
df_new
```

```python
df_new_sorted = df_new.sort_values(by='new_price', ascending=True).reset_index(drop=True)
```

```python
df_new_sorted
```

```python
sns.lineplot(df_new_sorted['new_price'])
```
